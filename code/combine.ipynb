```python
fault = ""

for year in range(1956, 2024):
    dfy = pd.read_csv("../processed data/" + str(year) + "/uitslagen" + str(year) + ".csv")

    dfy = dfy.replace("ADO", "ADO Den Haag")
    dfy = dfy.replace("AVC Heracles", "Heracles Almelo")
    dfy = dfy.replace("AZ '67", "AZ")
    dfy = dfy.replace("BVV Den Bosch", "FC Den Bosch")
    dfy = dfy.replace("Cambuur Leeuwarden", "SC Cambuur")
    dfy = dfy.replace("DS '79", "FC Dordrecht")
    dfy = dfy.replace("DWS/Amsterdam", "DWS")
    dfy = dfy.replace("Dordrecht '90", "FC Dordrecht")
    dfy = dfy.replace("EVV Eindhoven", "FC Eindhoven")
    dfy = dfy.replace("FC Den Bosch '67", "FC Den Bosch")
    dfy = dfy.replace("FC Den Haag", "ADO Den Haag")
    dfy = dfy.replace("FC Den Haag/ADO", "ADO Den Haag")
    dfy = dfy.replace("FC VVV", "VVV-Venlo")
    dfy = dfy.replace("FC Zwolle", "PEC Zwolle")
    dfy = dfy.replace("Feijenoord", "Feyenoord")
    dfy = dfy.replace("Fortuna Sittardia Combinatie", "Fortuna Sittard")
    dfy = dfy.replace("Go Ahead", "Go Ahead Eagles")
    dfy = dfy.replace("NAC", "NAC Breda")
    dfy = dfy.replace("PEC Zwolle '82", "PEC Zwolle")
    dfy = dfy.replace("RKC", "RKC Waalwijk")
    dfy = dfy.replace("RKSV Volendam", "FC Volendam")
    dfy = dfy.replace("RV & AV Excelsior", "Excelsior Rotterdam")
    dfy = dfy.replace("Roda JC", "Roda JC Kerkrade")
    dfy = dfy.replace("SBV Excelsior", "Excelsior Rotterdam")
    dfy = dfy.replace("SC Heracles '74", "Heracles Almelo")
    dfy = dfy.replace("SHS", "Holland Sport")
    dfy = dfy.replace("SVV/Dordrecht '90", "FC Dordrecht")
    dfy = dfy.replace("Sparta", "Sparta Rotterdam")
    dfy = dfy.replace("Sportclub Excelsior", "Excelsior Rotterdam")
    dfy = dfy.replace("VVV", "VVV-Venlo")
    dfy = dfy.replace("Wageningen", "FC Wageningen")
    dfy = dfy.replace("Xerxes", "Xerxes/DHC")

    dfy = dfy.replace("DWS", "AFC DWS")
    dfy = dfy.replace("Elinkwijk", "USV Elinkwijk")
    dfy = dfy.replace("Haarlem", "HFC Haarlem")
    dfy = dfy.replace("MVV", "MVV Maastricht")
    dfy = dfy.replace("NEC", "N.E.C.")
    dfy = dfy.replace("Veendam", "SC Veendam")

    clubs = list(dfy["Thuis"].unique())

    rows = []

    for club in clubs:
        rows.append([club, 0, 0, 0, 0, 0, 0, 0, 0])

    df2y = pd.DataFrame(rows, columns=["Club", "Gs", "W", "G", "V", "Pnt", "Dv", "Dt", "Ds"])

    for i in range(len(dfy)):
        game = dfy.iloc[i].to_dict()

        df2y["Gs"][df2y["Club"] == game["Thuis"]] += 1
        df2y["Gs"][df2y["Club"] == game["Uit"]] += 1

        df2y["Dv"][df2y["Club"] == game["Thuis"]] += game["Thuis Doelpunten"]
        df2y["Dt"][df2y["Club"] == game["Thuis"]] += game["Uit Doelpunten"]
        df2y["Dv"][df2y["Club"] == game["Uit"]] += game["Uit Doelpunten"]
        df2y["Dt"][df2y["Club"] == game["Uit"]] += game["Thuis Doelpunten"]

        if game["Uitslag"] == "Gewonnen":
            df2y["W"][df2y["Club"] == game["Thuis"]] += 1
            df2y["V"][df2y["Club"] == game["Uit"]] += 1
        elif game["Uitslag"] == "Gelijk":
            df2y["G"][df2y["Club"] == game["Thuis"]] += 1
            df2y["G"][df2y["Club"] == game["Uit"]] += 1
        else:
            df2y["V"][df2y["Club"] == game["Thuis"]] += 1
            df2y["W"][df2y["Club"] == game["Uit"]] += 1
        
    if year <= 1994:
        df2y["Pnt"] = 2 * df2y["W"] + df2y["G"]
    else:
        df2y["Pnt"] = 3 * df2y["W"] + df2y["G"]

    df2y["Ds"] = df2y["Dv"] - df2y["Dt"] 
    df2y = df2y.sort_values(by=["Pnt", "Ds"], ascending=False)

    # Define paths
    eindstand_dir = f"../processed data/{year}"
    file_path_eindstand= f"{eindstand_dir}/eindstand{year}.csv"
    file_path_eindstand_modern = f"{eindstand_dir}/eindstand_modern{year}.csv"

    # Ensure directories exist
    os.makedirs(eindstand_dir, exist_ok=True)

    # Check and resolve conflicts for processed file
    if os.path.isfile(file_path_eindstand):
        os.remove(file_path_eindstand)  # Remove the conflicting file
    elif os.path.isdir(file_path_eindstand):
        import shutil
        shutil.rmtree(file_path_eindstand)  # Remove the conflicting directory

    # Write processed CSV file
    df2y.to_csv(file_path_eindstand, index=False)

    df2y["Pnt"] = 3 * df2y["W"] + df2y["G"]
    df2y = df2y.sort_values(by=["Pnt", "Ds"], ascending=False)
    df2y.to_csv(file_path_eindstand_modern, index=False)

```
